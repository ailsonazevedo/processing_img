```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("beijaflor.jpg")

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img = cv.imread("beijaflor.jpg")

img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

plt.imshow(img)
plt.show()

cv.waitKey(0)
```
